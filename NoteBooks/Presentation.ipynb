# Point of Departure:  _The Age-Area Hypothesis_ 

"...the area that has the greatest diversity...is likely to be the location of the original homeland...In saying this, 
we are assuming the lowest number of population movements to account for the geographical distribution of the subgroups..." [Crowley and Bowern, 2010, p. 313](https://books.google.com/books?id=_N8v-s6fyt0C&pg=PA313&lpg=PA313&dq=historical+linguistics+age-area+hypothesis&source=bl&ots=O55FCvNw0O&sig=u-0kKlYDu_U2SUxIZhxOsYVdJhw&hl=en&sa=X&ved=0ahUKEwjT06-8oq_LAhXM1x4KHfEXAhAQ6AEIMDAC#v=onepage&q=historical%20linguistics%20age-area%20hypothesis&f=false)

Somewhat imprecisely stated - [Merritt Ruhlen's](http://www.amazon.com/The-Origin-Language-Tracing-Evolution/dp/0471159638) variation is point of origin is where the _most divergent_ language is - he credits it to [Edward Sapir](https://en.wikipedia.org/wiki/Edward_Sapir). 

# Basis for the AAH? 

## (Anthropological) Age-Area Hypothesis 
Attributed to [Clark Wissler]( https://en.wikipedia.org/wiki/Clark_Wissler#cite_note-Clark_Wissler_1999-3)- age of a cultural practice proportional to the breadth of its geographical distribution. Language a special case? 


## Maximum genetic diversity 

Maximum _genetic_ diversity occurs at the point of origin. A fact used to [support modern humans' origins in Africa](http://www.sciencedaily.com/releases/2009/04/090430144524.htm). Why? A sub-population is [not as diverse as the entire population](http://www.sciencedaily.com/releases/2007/07/070718140829.htm). 

# Questions and comments: 
    
* Minimum population movements does not by itself imply **AAH**, but suggests a plan of attack.
* The analogy with population biology does not work...People don't migrate with a fraction of language vocabularly.
* Lack of precision in statements like "likely."

# Building a Theory

## Back to my example: 

Why would one think that the events starting at $A$ is more likely? Occam's razor-like belief that this is simpler than the events starting at $C$ or any other point. More technically, this explanation **_conserves migratory events_**.

In some sense, we only need one migratory sequence to explain things in the $A$-based scenario. 

Question: how can this be made operational in terms of mathematics? Precedent: models associating least-cost algorithms (parsimony) to likelihood. [Felsenstein (2004)](http://www.amazon.com/Inferring-Phylogenies-Joseph-Felsenstein/dp/0878931775). 

# Underpinnings of a Model - basic assumptions

1. Migrations occur as random jumps.
2. Within a given sequence of migrations, jumps arrive according to a migratory-event-specific distribution. 
3. Once a jump has occurred from a location, the location goes dormant - subsequent migrations
   from this location require introduction of a new migratory sequence. 

# Narrowing assumptions:

1. Within a given migratory sequence, jumps occur through time according to an exponential distribution. _The parameter $\lambda$ is    specific to the sequence_.
2. Migrations are teleportations - ignore time and distance spent migrating for now (I use distance to resolve ties in my implementation). 

# Likelihood Comparison

## Migration starting at A: 
$
P(A;\lambda) = \lambda e^{-\lambda t_0}\lambda e^{-\lambda t_1} \lambda e^{-\lambda t_2} \lambda e^{-\lambda t_3} e^{-\lambda t_8}  = \lambda^4 e^{-\lambda( t_0+ t_1+ t_2+ t_3 +t_8)} 
$



## Migration starting at C:
$
P(C;\lambda_1, \lambda_2, \lambda_3) = \lambda_1 e^{-\lambda_1 (t_0+t_4)}\lambda_2 e^{-\lambda_2 (t_1+t_5)} \lambda_3 e^{-\lambda t_2} \lambda_3 e^{-\lambda t_3} e^{-\lambda_3 t_8}  = \lambda_1 e^{-\lambda_1 (t_0+t_4)}\lambda_2 e^{-\lambda_2 (t_1+t_5)} \lambda^2_3 e^{-\lambda_2( t_2+ t_3+ t_8)}
$


# In Logs:

## A:
$
LL_(A;\lambda) = 4 \ln \lambda-\lambda( t_0+ t_1+ t_2+ t_3 +t_8) 
$

## B:
$
LL_(B;\lambda_1,\lambda_2,\lambda_3) = \ln \lambda_1 -\lambda_1 (t_0+t_4) + \ln \lambda_2 -\lambda_2 (t_1+t_5) - 2 \ln \lambda_3 -\lambda_2( t_2+ t_3+ t_8)
$


Maximum likelihood estimate of $\lambda$ is:

$
\lambda^* = \frac{n}{\sum_{i=1}^n t_i}
$

# Maximized w.r.t. Parameters:

$
LL_A = 4 (\ln 4 - 1) - 4 \ln (t_0+ t_1+ t_2+ t_3 +t_8)
$

$
LL_C = 2 (\ln 2 - 2) - \ln (t_0 + t_4) - \ln (t_1 +t _5) - 2 \ln (t_2+t_3+t_8)
$

** Likelihood of each route being "the" route. **

Suppose that $\Delta$ is the size of a unit of time, with $ \Delta = \delta T$, and $\delta = \frac{1}{5}$. 

## Comparison boils down to:
$
4 \ln 4 > 2 \ln 2 - \ln (1-\delta)(1-\delta)^2
$ or
$ 4^4 > \frac{2^2}{(1-\delta)(1-\delta)^2} $

Which becomes $ (1-\delta)(1-2\delta)^2 > \frac{1}{2^6}$. 

** Usually not satisfied for "reasonable" $\delta$, for example:if $\delta = 1/5$, $LHS = .48$, $RHS = .016$

## Some more Intuition 

* Note how the number of branches that are included, negatively, in the likelihood changes with more complex explanations.
* Also, note how multiple migration rules have to "stretch" the parameter to explain an early jump...and then nothing!
* Jensen's inequality

# An unconditional model: 
So far, comparison of entities like $P(A|m=1)$ and $P(C|m=3)$. 

A better model compare $P(A \cap (m = 1))$ and $P(C \cap (m=3))$? 

Form a model of $P(m=M)$.

## Accordingly: 
Suppose that migratory sequences start at random times according to a Poisson distribution with parameter $\rho$. Then: 

$
P(m=M) = \frac{(\rho T)^{M}e^{-\rho T}}{M!}
$



# Maximize w.r.t. Parameters:

$ \rho^*(M)=\frac{M}{T}$. 

Then: 
$
P(m=M) = \frac{M^{M}e^{-M}}{M!}  = \frac{M^{M}e^{-M}}{M(M-1)!}
$

Since $(M-1)! \approx \sqrt{\frac{2\pi}{M}}\frac{M^M}{e^M}$


$
P(m=M) \approx \sqrt{\frac{1}{2\pi M}}
$


Or in logs:
$
-\frac{1}{2}\left(\ln(2\pi) + \ln M\right)
$

In some ways, a better first cut, because the relative probability is simply inversely proportional to $M$!

** Direct relation to parsimony **

# Dynamic Programming Algorithm for calculation

## Idea:
We keep track of a state vector for each location: $S_i= \{A_i,N_i,T_i,LL_i\}$. 
* $A_i$ is equal to either `True` or `False`. The value is assigned to `True` if any migration has started from this point.
* $N_i$ is the number of migrations following from this point.
* $T_i$ is the branch length going forward. 
* $LL_i$ is a continuation likelihood.

** Intuition ** - work backwards through the tree calculating the probability of everything from the current point on started at each point. 

# Iteration (backwards through Tree)

1. Start with $S_i =\{A_i = \texttt{False}, N_i = 0, T_i = t_i, LL_i = 0\}$, $i=1,2,3,...,N$
2. Pick a node and get the branch leading to the node $t_l$ and $S_j$ for groups $J$ above the node. 
3. For each $j \in J$:
   1. If inactive, for each $k$ in $J \setminus j$:
       1. increment $N_k$ by one.
       2. Add $t_l+t_k$ to $T_j$, compute $LL_k$, and clear $N_i$ and $T_i$.
       3. Pick $k^*$ as $\max_k {LL'_k}$
       3. Set $A_i = \texttt{True}$, $N_i=N_k$, and add $LL_i = LL_i+LL_{k^*}$
   2. If active:
       1. Use $N_j$ and $T_J$ to compute log-likelihood and add result to $LL_j$. Set $N_j=1$ and set $T_j$ = $t_j+t_k^*$. 
4. Repeat 2-3 until there are no more nodes, and then follow B). 


** Result:** $\ln P(i|m)$. 

** Note ** $P(m)$ can be added after by just counting migrations needed.

# Integration with other methods - a Quick Preview

So, the idea is to recover who was where when. Accordingly, form distributions over locations and times, using:

1. Maximum likelihood location and corresponding likelihood value, 
2. Prior knowledge of language splits and extinction dates, with:
3. Likelihood-based "Nucleotide substitution" model of linguistic evolution.

$
P(T,L,\theta_s|D) = P(T|D_s,D_e\theta_s)P(T|D_w,\theta_w)P(L|T,D_w)
$

That is, the distribution of locations, times, (and parameters), implied by probabilistic models of locations, word drift, and what is known about when splits and extinctions occurred?

Phylogenetic analysis of languages usually relies on [**_Swadesh Lists_**](https://en.wikipedia.org/wiki/Swadesh_list). Idea: view base vocabularly and similarity judgements as a sort of genetic code for a culture. 

Word lists should evolve randomly - words very resistant to borrowing.

* Model language change over time using a model of genetic drift

```python
%matplotlib
matplotlib.rcParams['figure.figsize'] = (8, 6)

# Let's just define a base tree and then see if we can plant stuff on top of that. 
Num = np.matrix(np.arange(0, PyIETools.rows(Tree))).T
names = np.array(['A','B','C','D','E'])

minPos = PyIETools.J(PyIETools.rows(Tree), PyIETools.cols(Tree), np.nan)
maxPos = PyIETools.J(PyIETools.rows(Tree), PyIETools.cols(Tree), np.nan)
avePos = PyIETools.J(PyIETools.rows(Tree), PyIETools.cols(Tree), np.nan)
            
# Roll through the tree and positions point and lines accordingly.

for i in range(0, PyIETools.cols(Tree)):
    m=PyIETools.ps(Tree, i)
    for j in range(0, PyIETools.rows(m)):
        bracketed = PyIETools.psm(Num, m, j)
        lowest = min(bracketed)
        highest = max(bracketed)
        minPos[m[j,0]:m[j,1]+1,i:i+1] = PyIETools.J(1 + m[j,1] - m[j,0], 1, lowest)
        maxPos[m[j,0]:m[j,1]+1,i:i+1] = PyIETools.J(1 + m[j,1] - m[j,0], 1, highest)
        average = np.mean(bracketed)
        avePos[m[j,0]:m[j,1]+1,i:i+1] = PyIETools.J(1 + m[j,1] - m[j,0], 1, average)
# Block of code to to compile tree running sums. 
            
runningsum = np.copy(TFR[:,0:1])
for i in range(1,PyIETools.cols(TFR)):
    runningsum=np.hstack((runningsum, np.copy(np.matrix(np.nansum(TFR[:,:i+1],axis=1)).T)))
        
        # Reorganize the numbers so that the x-axis is the depth of the tree
            
depth = np.max(runningsum)
runningsum = runningsum - depth

        # Plot the root of the tree...

begin = [-depth,-depth + TFR[0,0]]
end   = [(PyIETools.rows(TFR) - 1)/2,(PyIETools.rows(TFR) - 1)/2]
plt.plot(begin, end)
plt.annotate('t0', xy = ( (begin[0]+(begin[1]-begin[0])/2),(end[0]+(end[1]-end[0])/2) ) )

count = 1
coverdlist = []
for c in range(0, PyIETools.cols(TFR) - 1):
    for r in range(0, PyIETools.rows(Tree)):
        y = [avePos[r,c], avePos[r,c+1]]
        x = [runningsum[r,c], runningsum[r,c+1]]
        plt.plot(x, y)
        apoint = x[0]+(x[1]-x[0])/2, y[0]+(y[1]-y[0])/2        
        if (x[0] != x[1]) and apoint not in coverdlist:
            text = 't' + str(count)
            count += 1
            plt.annotate(text, xy=apoint, textcoords = "offset points")
            coverdlist.append(apoint)

plt.axis([-depth - 1,np.max(runningsum) + 1.5,-1,PyIETools.rows(Tree) + 1],yticks='none')
     
Names = list(names)    
for n in range(0, len(Names)):
    plt.annotate(Names[n][0], xy = (runningsum[n,-1] + .2, n))  

plt.axis('off')

cwd = os.getcwd()

if 'Writeup' in cwd:
    pass
else:
    os.chdir(cwd+'\\Writeup')

plt.savefig('Basefig')
plt.close()

# Add some space to the right - not very mat plot libby, but it should work.
im1 = ImagePaste.open('Basefig.png')

blank_image = ImagePaste.new("RGB", (im1.size[0]*2,im1.size[1]), color = 'white')
blank_image.paste(im1,(0,0))

blank_image.save('Basefig.png')


```

```python
# Some stuff to string observations like a genetic code:
def genclass(word):
    Dog=['pbf', 't8dT!', 'szSZ', 'cjgkqGX7Cx', 'm', 'Nn45',
         'lLr', 'wv', 'y', 'ieE3auoh']
    if len(re.sub(' ','',word)) ==0:
        return '-'
    firstlet=word[0]
    
    if firstlet=='p' or firstlet=='b' or firstlet=='f':
        return u'p'
    elif firstlet=='t' or firstlet=='8' or firstlet=='d' \
            or firstlet=='T' or firstlet=='!':
        return u't'
    elif firstlet=='s' or firstlet=='z' or firstlet=='S' \
            or firstlet=='Z':
        return u's'
    elif firstlet=='c' or firstlet=='j' or firstlet=='g' \
            or firstlet=='k' or firstlet=='q' or firstlet=='G' \
            or firstlet=='X' or firstlet=='7' or firstlet=='C' \
            or firstlet=='x':
        return u'c'
    elif firstlet=='m':
        return u'm'
    elif firstlet=='N' or firstlet=='n' or firstlet=='4' or firstlet=='5':
        return u'N'
    elif firstlet=='l' or firstlet=='L' or firstlet=='r':
        return u'l'
    elif firstlet=='w' or firstlet=='v':
        return u'w'
    elif firstlet=='y':
        return u'y'
    elif firstlet=='i' or firstlet=='e' or firstlet=='a' or firstlet=='E' \
            or firstlet=='3' or firstlet=='u' or firstlet=='o' or firstlet=='h':
        return u'i'
    else:
        return u'z'

```
